```python
dt=df[['CategoryCode', 'ItemCode','dayofyear','DailySales']]
```

```python
dt
```

```python
id_list=dt['ItemCode'].unique() #get a list of unique item codes
```

```python
id_list
```

```python
final={}
for id in id_list:
    X_test=[]
    dt_item=dt[dt['ItemCode']==id]
    X_train= dt_item[['dayofyear']]
    Y_train=dt_item['DailySales']
    for i in range(137,165):
        X_test.append([i])
    from sklearn.linear_model import Lasso
    model =Lasso(normalize=True)
    model.fit(X_train, Y_train)
    y_test_predicted = model.predict(X_test)
    result=y_test_predicted.round(0)
    final[id]=result
```

```python
final
```

```python
final_result={}
for item in final:
    week_details=[]
    week_details.append(sum(final[item][0:7]).round(0))
    week_details.append(sum(final[item][7:14]).round(0))
    week_details.append(sum(final[item][14:21]).round(0))
    week_details.append(sum(final[item][21:28]).round(0))
    final_result[item]=week_details
```

```python
def updatesales(code,week):
    week_no=int(week[1])
    new_sales=final_result[code][week_no-1]
    print(new_sales)
    return int(new_sales)
sales_list=[]
for i in range(377):
    row=test_data.iloc[i]
    code=row[1]
    week=row[2]
    
    new_sale=updatesales(code,week)
    sales_list.append(new_sale)
test_data['PredictedSales']=sales_list
```
